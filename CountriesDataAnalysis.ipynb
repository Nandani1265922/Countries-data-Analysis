```python
import numpy as np
import pandas as pd

# Load the dataset ( CSV file  should be in the same folder)
df = pd.read_csv("countries of the world.csv", decimal=',')

# --- Data Exploration ---
print("Dataset Shape:", df.shape)
print("\nColumns:", df.columns)
print("\nDataset Info:")
df.info()
print("\nStatistical Summary:")
print(df.describe())

# --- Data Preprocessing ---
# Remove extra spaces from the 'Region' column
df['Region'] = df['Region'].str.strip()

# Convert 'Literacy (%)' to numeric type, handling errors gracefully
df['Literacy (%)'] = pd.to_numeric(df['Literacy (%)'], errors='coerce')
# 1. Country with the highest population
highest_pop_country = df[df['Population'] == df['Population'].max()]['Country'].values[0]
print(f"\nCountry with the highest population: {highest_pop_country}")

# 2. Country with the highest literacy rate
highest_literacy_country = df[df['Literacy (%)'] == df['Literacy (%)'].max()]['Country'].values[0]
print(f"Country with the highest literacy rate: {highest_literacy_country}")

# 3. Sort countries by population (Highest to Lowest)
df.sort_values(by='Population', ascending=False, inplace=True)
print("\nTop 5 Most Populated Countries:\n", df['Country'].head())

# 4. Total countries per region
print("\nTotal countries per region:\n", df['Region'].value_counts())

# 5. Count of countries in 'NEAR EAST' region
near_east_count = df['Region'].value_counts().get('NEAR EAST', 0)
print(f"\nNumber of countries in 'NEAR EAST': {near_east_count}")

# 6. Filter for BALTICS region
baltics_df = df[df['Region'] == 'BALTICS']

# 7. Second largest populated country overall
try:
    second_largest_country = df.sort_values(by='Population', ascending=False).iloc[1]['Country']
    print(f"\nSecond largest populated country: {second_largest_country}")
except IndexError:
    pass

# 8. Missing values checks for critical fields
print("\nMissing values in Agriculture by Country:\n", df[df['Agriculture'].isna()]['Country'])
print("Missing values in Industry by Country:\n", df[df['Industry'].isna()]['Country'])
print("Missing values in Service by Country:\n", df[df['Service'].isna()]['Country'])

# 9. Country with highest population in BALTICS region
if not baltics_df.empty:
    highest_pop_baltics = baltics_df[baltics_df['Population'] == baltics_df['Population'].max()]
    print(f"\nHighest populated country in BALTICS: {highest_pop_baltics['Country'].values[0]}")
    print(f"Population: {highest_pop_baltics['Population'].values[0]}")
```
